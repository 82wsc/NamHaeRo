```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager, rc
```

```python
# 적용인구 현황
inf = pd.read_excel("02_시군구별 제2권-대전 충북 충남 세종_2019년.xlsx",sheet_name='82p')

# 1. 3번 행을 칼럼명으로 변경
inf.columns = inf.iloc[3]

# 2. 0, 1, 2, 4번 행 삭제
inf = inf.drop(index=[0, 1, 2, 3, 4, 5, 6, 7])

# 3. 46, 47, 48번 행 삭제
inf = inf.drop(index=[47, 48, 49])

# 인덱스 리셋
inf = inf.reset_index(drop=True)

inf.iloc[:, [2, 3]] = inf.iloc[:, [3, 2]].values
inf.iloc[:, [6, 7]] = inf.iloc[:, [7, 6]].values

inf = inf.loc[:, ~inf.columns.isna()]

inf = inf.drop(columns = ["G. & T.", "Division"])

inf.columns = [col.split('\n')[0] for col in inf.columns]
inf.columns = [col.split('  C.E')[0] for col in inf.columns]
inf.columns = [col.split('  S.E.')[0] for col in inf.columns]
inf.columns = [col.split('    Medical Aid')[0] for col in inf.columns]

# 창원시 관련 행 필터링 및 합계 계산
chungju =[ '청주시상당구             ','청주시서원구             ', '청주시흥덕구             ', '청주시청원구             ']
chungju_data = inf[inf['구  분'].isin(chungju)].iloc[:, 1:].sum()

# 새로운 창원시 행 생성
chungju_row = pd.DataFrame([['청주시'] + chungju_data.tolist()], columns=inf.columns)

# 전주시 행 추가 및 기존 구 삭제
inf = pd.concat([inf[~inf['구  분'].isin(chungju)], chungju_row], ignore_index=True)

# 창원시 관련 행 필터링 및 합계 계산
chunan =[ '천안시동남구', '천안시서북구']
chunan_data = inf[inf['구  분'].isin(chunan)].iloc[:, 1:].sum()

# 새로운 창원시 행 생성
chunan_row = pd.DataFrame([['천안시'] + chunan_data.tolist()], columns=inf.columns)

# 전주시 행 추가 및 기존 구 삭제
inf = pd.concat([inf[~inf['구  분'].isin(chunan)], chunan_row], ignore_index=True)
inf = inf.rename(columns={'구  분': '지역'})


inf['지역'] = inf['지역'] + ' ' + inf.groupby('지역').cumcount().replace(0, '').astype(str)

inf
```
