#남해, 통영, 여수 의료 인프라 비교

```python
# CSV 파일 불러오기
file_path_relative = "남해군_기준_의료인프라_상대평가.csv"
df_relative = pd.read_csv(file_path_relative)

# 데이터 시각화 준비
categories_relative = ["병상 수", "의료기관 수", "관내 이용 비율", "진료 과목 다양성", "고령화율"]
regions = df_relative["시군구"]

# 상대 평가 데이터 추출
relative_columns = [f"{col}_비교" for col in categories_relative]
relative_data = df_relative[relative_columns].values

# 남해군 기준 상대 평가 시각화
fig, ax = plt.subplots(figsize=(10, 6))
x = np.arange(len(relative_columns))
width = 0.25

for i, region in enumerate(regions):
    ax.bar(x + i * width, relative_data[i], width, label=region)

ax.set_xlabel("평가 지표")
ax.set_ylabel("남해군 대비 상대적 점수")
ax.set_title("남해군 기준 시군구별 의료 인프라 상대 평가")
ax.set_xticks(x + width / 2)
ax.set_xticklabels(categories_relative, rotation=45)
ax.legend(title="시군구")
plt.tight_layout()
plt.show()
```

# 통영-남해 의료 인프라 세부비교

```python
# 통영-남해군 의료 비교 평가
file_path_comparison = "통영_남해_의료비교평가_세부데이터.csv"
comparison_data = pd.read_csv(file_path_comparison)

categories = comparison_data["항목"]
namhae_data = comparison_data["남해군"]
tongyeong_data = comparison_data["통영시"]

# 막대 위치와 너비 설정
x = np.arange(len(categories))
width = 0.35

# 남해군 vs 통영시 비교 시각화
fig, ax = plt.subplots(figsize=(12, 8))
ax.bar(x - width / 2, namhae_data, width, label="남해군")
ax.bar(x + width / 2, tongyeong_data, width, label="통영시")

ax.set_xlabel("항목")
ax.set_ylabel("값")
ax.set_title("남해군 vs 통영시 의료 비교 평가")
ax.set_xticks(x)
ax.set_xticklabels(categories, rotation=45, ha="right")
ax.legend(title="지역")
plt.tight_layout()
plt.show()
```

```python
# 항목별 남해군-통영시 비교 개별 시각화
for i, category in enumerate(categories):
    fig, ax = plt.subplots(figsize=(6, 4))
    ax.bar(["남해군", "통영시"], [namhae_data[i], tongyeong_data[i]], color=['blue', 'orange'])
    ax.set_title(f"{category} 비교")
    ax.set_ylabel("값")
    ax.set_xlabel("지역")
    plt.tight_layout()
    plt.show()
```
