### 관광활성화지수

```python
import os
import unicodedata

# 관광활성화지수 파일들이 저장된 경로
directory_path = '/content/관광활성화 세부지표'

# 해당 디렉토리에서 모든 CSV 파일 리스트 가져오기
csv_files = [file for file in os.listdir(directory_path) if file.endswith('.csv')]

# 빈 리스트 생성
dataframes = []

# 각 CSV 파일을 데이터프레임으로 읽어와 리스트에 추가
for file in csv_files:
    file_path = os.path.join(directory_path, file)
    try:
        # UTF-8로 읽기 시도
        temp = pd.read_csv(file_path, encoding='utf-8')
    except UnicodeDecodeError:
        try:
            # UTF-8 실패 시 euc-kr로 읽기 시도
            temp = pd.read_csv(file_path, encoding='euc-kr')
        except UnicodeDecodeError:
            # 그래도 실패하면 cp949로 읽기 시도
            temp = pd.read_csv(file_path, encoding='cp949')

    # 파일 이름에서 지역명을 추출하여 컬럼 추가
    region_parts = os.path.basename(file).split('_')[0:2]  # 시도명과 시군구명 추출
    region_name = ' '.join(region_parts)  # 시도명과 시군구명을 공백으로 연결

    # 유니코드 정규화
    region_name = unicodedata.normalize('NFC', region_name)
    temp['지역명'] = region_name

    dataframes.append(temp)

# 모든 데이터프레임을 세로로 병합
df4 = pd.concat(dataframes, ignore_index=True)

# 유니코드 정규화 적용
df4['지역명'] = df4['지역명'].apply(lambda x: unicodedata.normalize('NFC', x))

# 제주도 지역 제외
temp = df4[df4['지역명'].str.startswith('제주')]
df4 = df4.drop(temp.index, axis=0)

# 병합된 데이터프레임 확인
display(df4)
```

```python
df4['지역명'].unique()
```

```python
# 지역명 기준 병합을 위한 처리
df4['지역명'] = df4['지역명'].str.strip()
df4['지역명'] = df4['지역명'].str.encode('utf-8').str.decode('utf-8')

import unicodedata

# 유니코드 정규화 적용
df4['지역명'] = df4['지역명'].apply(lambda x: unicodedata.normalize('NFC', x))

df5 = df4.groupby('지역명', as_index=False)['관광활성화 지수'].mean().round(2)
df5
```

### 숙박시설 수

```python
import os
import zipfile
import chardet  # 인코딩 감지를 위한 라이브러리
from io import BytesIO  # Python 표준 라이브러리 사용

# 파일들이 저장된 경로를 지정하세요
directory_path = '/content/숙박업소 현황'

# 해당 디렉토리에서 모든 zip 파일 리스트 가져오기
zip_files = [file for file in os.listdir(directory_path) if file.endswith('.zip')]

# 빈 리스트 생성
dataframes = []

# 각 zip 파일 처리
for zip_file in zip_files:
    zip_file_path = os.path.join(directory_path, zip_file)

    # zip 파일 열기
    with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
        namelist = zip_ref.namelist()
        if not namelist:  # zip 파일이 비어 있는 경우
            print(f"빈 zip 파일입니다: {zip_file}")
            continue

        # zip 파일 안의 첫 번째 파일 (csv 파일) 이름 가져오기
        csv_filename = namelist[0]

        # zip 파일에서 csv 파일 추출 후 읽기
        with zip_ref.open(csv_filename) as csv_file:
            # 파일 내용 읽기
            raw_data = csv_file.read()

            # 인코딩 감지
            encoding_result = chardet.detect(raw_data)
            detected_encoding = encoding_result['encoding']

            try:
                # 감지된 인코딩으로 읽기
                temp = pd.read_csv(BytesIO(raw_data), encoding=detected_encoding)
            except Exception as e:
                print(f"파일을 읽을 수 없습니다: {csv_filename}, 오류: {e}")
                continue

    # 파일명에서 지역명 추출
    try:
        region_name = zip_file.split('_')[1].replace('+', ' ')
    except IndexError:
        print(f"지역명을 추출할 수 없습니다: {zip_file}")
        continue

    # 데이터프레임에 지역명 열 추가
    temp['지역명'] = region_name

    # 데이터프레임 리스트에 추가
    dataframes.append(temp)

# 모든 데이터프레임을 세로로 병합
if dataframes:  # 비어 있는 경우를 방지
    df7 = pd.concat(dataframes, ignore_index=True)
else:
    print("유효한 데이터를 포함한 파일이 없습니다.")

# 유니코드 정규화 적용
df7['지역명'] = df7['지역명'].apply(lambda x: unicodedata.normalize('NFC', x))

# 제주도 지역 제외
temp = df7[df7['지역명'].str.startswith('제주')]
df7 = df7.drop(temp.index, axis=0)

# 데이터프레임 확인
display(df7)
```

```python
df7['지역명'].unique()
```

```python
# 기준년도, 업종명 drop
df7 = df7.drop(['기준년도','업종명'], axis=1)
display(df7)
```

```python
# groupby 하여 merge
df7 = df7.rename(columns={'객실현황수' : '숙박객실수'})
df7 = df7.groupby('지역명', as_index=False)['숙박객실수'].sum()

merged_df4 = pd.merge(merged_df3, df7, on='지역명', how='left')
merged_df4
```

```python
merged_df4.loc[merged_df4.isna()['숙박객실수']==True]
# 부천시 객실수 데이터 없음
```

```python
# 부천의 객실 수 추론
# 부천시 인구수와 관광활성화 지수 설정
bucheon_population = 7357360  # 부천시 인구수
bucheon_tourism_activation_index = 81.71  # 부천시 관광활성화 지수

# 유사한 지역 필터링
similar_regions = merged_df4[
    (merged_df4['관광활성화 지수'] >= bucheon_tourism_activation_index * 0.8) &
    (merged_df4['관광활성화 지수'] <= bucheon_tourism_activation_index * 1.2)
]

print(similar_regions)

# 유사 지역들의 숙박객실 수 평균 계산
estimated_rooms = similar_regions['숙박객실수'].mean().round(0)
print(f"부천시의 추정 숙박객실 수: {estimated_rooms:.2f}")
```
